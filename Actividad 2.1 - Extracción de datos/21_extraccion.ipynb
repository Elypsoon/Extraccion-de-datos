```python
import pandas as pd
```

```python
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
# Filtramos específicamente por el nombre del vendedor
filtro1 = df[df['NOMBRE_VENDEDOR'] == 'LETICIA RAMIREZ HERNANDEZ']
# Lo pasamos a un csv
filtro1.to_csv('filtro1.csv')
filtro1
```

```python
# Filtramos las filas intermedias
filtro2 = df.iloc[800:1000, :]
filtro2.to_csv('filtro2.csv')
filtro2
```

```python
# Filtramos filas específicas
filtro2_1 = df.iloc[[10, 12, 20, 100, 500], :]
filtro2_1.to_csv('filtro2_1.csv')
filtro2_1
```

```python
# Filtramos por columnas consecutivas
filtro3 = df.iloc[:, 0:5]
filtro3.to_csv('filtro3.csv')
filtro3
```

```python
# Filtramos por columnas específicas
filtro3_1 = df.iloc[:, [0, 2, 4, 6, 8]]
filtro3_1.to_csv('filtro3_1.csv')
filtro3_1
```

```python
# Hacemos otro CSV con un índice específico
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
# Filtro de filas vs columnas
filtro4 = df1.loc[["AARON NUÑO CORDOVA", "DIEGO ARMANDO LEON DURAN"], ["NOMBRE_CLIENTE"]]
filtro4.to_csv('filtro4.csv')
filtro4
```

```python
# Filtro por cabecera
filtro5 = df.head(10)
filtro5.to_csv('filtro5.csv')
filtro5
```

```python
# Filtro por cola
filtro6 = df.tail(10)
filtro6.to_csv('filtro6.csv')
filtro6
```

```python
# Filtro por comparación
# Lo hacemos en base al costo total calculado
filtro7 = df[df['COSTO_TOTAL_CALCULADO'] > 1000]
filtro7.to_csv('filtro7.csv')
filtro7
```

```python
# Filtro lógico Y (AND)
filtro8 = df[(df['COSTO_TOTAL_CALCULADO'] > 1000) & (df['CANT'] > 10)]
filtro8.to_csv('filtro8.csv')
filtro8
```

```python
# Filtro lógico O (OR)
filtro9 = df[(df['COSTO_TOTAL_CALCULADO'] < 1000) | (df['CANT'] > 10)]
filtro9.to_csv('filtro9.csv')
filtro9
```

```python
# Filtro lógico de negación (NOT)
# Quitamos a todos los que son el público en general
filtro10 = df[~(df['NOMBRE_CLIENTE'] == 'PÚBLICO EN GENERAL')]
filtro10.to_csv('filtro10.csv')
filtro10
```
