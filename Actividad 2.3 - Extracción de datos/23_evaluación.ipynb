```python
import pandas as pd
```

```python
# Cargamos la base de datos
df = pd.read_csv('listings - New York.csv')
df
```

```python
# a) Primer filtro: 5 registros de host_name diferentes
# Aqu√≠ se seleccionan los registros que contienen los nombres de los hosts Anna, Obed, Michael, Edward y Jennifer
filtro1 = df[df['host_name'].isin(['Anna', 'Obed', 'Michael', 'Edward', 'Jennifer'])]
filtro1.to_csv('filtro1.csv')
filtro1
```

```python
# b) Segundo filtro: los registros de los host que se hayan unido a Airbnb antes de 2019
# Para este filtro, usaremos la columna 'host_since' que contiene la fecha en la que el host se uni√≥ a Airbnb
filtro2 = df[df['host_since'] < '2019-01-01']
filtro2.to_csv('filtro2.csv')
filtro2
```

```python
# c) Tercer filtro: los registros de los host que que responden a m√°s tardar un d√≠a
# Para este filtro, usaremos la columna 'host_response_time' que contiene el tiempo de respuesta del host
filtro3 = df[df['host_response_time'] == 'within a day']
filtro3.to_csv('filtro3.csv')
filtro3
```

```python
# d) Cuarto filtro: los registros que tengan al menos 5 accommodates
# Para este filtro, usaremos la columna 'accommodates' que contiene el n√∫mero de personas que pueden ser acomodadas
filtro4 = df[df['accommodates'] >= 5]
filtro4.to_csv('filtro4.csv')
filtro4
```

```python
# e) Quinto filtro: los registros de los tipos de cuarto 'Entire home/apt' y 'Private room'
# Para este filtro, usaremos la columna 'room_type' que contiene el tipo de cuarto
filtro5 = df[df['room_type'].isin(['Entire home/apt', 'Private room'])]
filtro5.to_csv('filtro5.csv')
filtro5
```

```python
# f) Sexto filtro: los registros que a lo mucho cuenten con 4 camas
# Para este filtro, usaremos la columna 'beds' que contiene el n√∫mero de camas
filtro6 = df[df['beds'] <= 4]   
filtro6.to_csv('filtro6.csv')
filtro6
```

```python
# g) S√©ptimo filtro: los registros que tengan 1 ba√±o o sean superhosts pero con la condici√≥n de que todos sean instant bookable
# Para este filtro, usaremos las columnas 'bathrooms', 'host_is_superhost' y 'instant_bookable'
filtro7 = df[((df['bathrooms'] == 1) | (df['host_is_superhost'] == 't')) & (df['instant_bookable'] == 't')]
filtro7.to_csv('filtro7.csv')
filtro7
```

```python
# h) Octavo filtro: los registros cuyo review score cleanliness sea menor a 3 y review score rating sea mayor a 3
# Para este filtro, usaremos las columnas 'review_scores_cleanliness' y 'review_scores_rating'
filtro8 = df[(df['review_scores_cleanliness'] < 3) & (df['review_scores_rating'] > 3)]
filtro8.to_csv('filtro8.csv')
filtro8
```
